```python
#Importamos la libreria a ocupar
import pandas as pd
```

```python
#Ruta de origen del dataset
ruta = r"C:\Users\Elias\Documents\Primer Proyecto Individual\Datasets limpios\games_clean"
#Cargamos el dataset a trabajar
df_games_clean = pd.read_csv(ruta)
df_games = df_games_clean
df_games.head()
```

```python
genres_dummies = (
    df_games['genres'].explode()
    .str.get_dummies()
    .groupby(level=0)
    .sum()
    .reset_index(drop=True)
)

generos_totales = pd.concat([df_games[['id']], genres_dummies], axis=1)


keep_cols = generos_totales.columns[~generos_totales.columns.str.contains('&amp;', case=False)]

generos_totales= generos_totales[keep_cols]

generos_totales.columns = pd.unique(generos_totales.columns)

generos_totales
```

```python
#Creamos un nuevo dataset en formato parquet
generos_totales.to_parquet(r"C:\Users\Elias\Documents\Primer Proyecto Individual\Datasets limpios\games_clean_dummies",index= False)
```
